```python
bins = [0, 100000, 200000, 300000, 400000, 600000]
labels = ['0-100000','100001-200000','200001-300000','300001-400000','400000+']

df['Income Band'] = pd.cut(
    df['Estimated Income'],
    bins=bins,
    labels=labels,
    right=True   # [inclusive, exclusive)
)


```

```python
# Examine the distribution of unique categories in  categorical columns

categorical_cols = df[['BRId', 'GenderId', 'IAId','Amount of Credit Cards','Nationality','Occupation', 'Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']].columns

for col in categorical_cols:
    print(f"Value Counts for {col} : ")
    display(df[col].value_counts())
```

```python
import matplotlib.pyplot as plt

for col in categorical_cols:
    vc = df[col].value_counts().sort_index()   

    plt.figure()
    plt.bar(vc.index.astype(str), vc.values)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.title(f"Bar Chart of {col} (Category Order)")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for i, predictor in enumerate(categorical_cols):
    plt.figure()
    sns.countplot(data=df, x=predictor, hue='GenderId', order=sorted(df[predictor].unique()))
    plt.show()
```

```python
for i, predictor in enumerate(categorical_cols):
    plt.figure()
    sns.countplot(data=df, x=predictor, hue='Nationality', order=sorted(df[predictor].unique()))
    plt.show()
```

```python
# histplot of value counts for different Occupations

occ_counts = df['Occupation'].value_counts()
sns.histplot(occ_counts)
plt.xlabel('Number of Customers per Occupation')
plt.ylabel('Number of Occupations')
plt.title('Distribution of Occupation Frequencies')
plt.show()
```

```python
# Numerical Analysis
numerical_cols = ['Estimated Income', 'Superannuation Savings','Credit Card Balance', 'Bank Loans',
       'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
       'Foreign Currency Account', 'Business Lending']

# univariate analysis
plt.figure(figsize=(15,10))
for i,col in enumerate(numerical_cols):
    plt.figure(figsize=(10,10))
    sns.histplot(df[col],kde=True)
    plt.title(col)
    plt.show()

```
